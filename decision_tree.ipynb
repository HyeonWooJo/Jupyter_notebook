```python
import matplotlib.pyplot as plt
```

```python
from sklearn import tree
```

```python
X = [[0,0], [1,1]]
Y = [0,1]
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
clf = clf.fit(X,Y)
```

```python
clf
```

```python
clf.predict([[2,2]])
```

- Y값이 0 또는 1뿐이라 0, 1 중 하나로 답이 나옴

```python
clf.predict_proba([[2,2]])
```

- predcit_proba는 값이 비율로 출력됨

```python
from sklearn.datasets import load_iris
# 데이터셋이 라이브러리에 내장되어 있는 경우임
# Decision tree 알고리즘은 숫자만 이해할 수 있기 때문에 0,1,2로 카테고리 구분
```

```python
X, y = load_iris(return_X_y =True)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
import graphviz
plt.figure(figsize=(17,10))
t = tree.plot_tree(clf.fit(X,y), filled=True)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render('iris')
```
